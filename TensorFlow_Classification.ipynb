```python
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
def sigmoid(X):
  return 1 / (1 + tf.exp(-X))

def relu(X):
  return tf.maximum(0, X)
```

```python
X = tf.cast(tf.range(-10, 10), tf.float32)
X
```

```python
relu(X)
```

```python
sigmoid(X)
```

```python
plt.plot(sigmoid(X))
```

```python
plt.plot(relu(X))
```

```python
y = relu(X) + sigmoid(X)
y
```

```python
plt.plot(y)
```

```python
from sklearn.datasets import make_circles

# number of samples
n_samples = 1000

# Create Circles
X, y = make_circles(n_samples,
                    noise=0.03,
                    random_state=42)
```
